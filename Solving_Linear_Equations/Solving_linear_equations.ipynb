# Solving System of Equations

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from mpl_toolkits.mplot3d import Axes3D
from IPython import display
import time
from sympy import *
```

```python
# Equation form: ay = bx + c
# a, b, c: coefficients

eq1 = np.array([1, 2, 1])
eq2 = np.array([2, 1, 4])


# plotting the solutions
# using values of x from -5 to 5
y = ([eq1[1] * -5, eq1[1] * 5] + eq1[2]) / eq1[0]    # solution for eq1
plt.plot([-5, 5], y)

# using values of x from -5 to 5
y = ([eq2[1] * -5, eq2[1] * 5] + eq2[2]) / eq2[0]    # solution for eq2
plt.plot([-5, 5], y)


plt.axis([-5, 5, -5, 5])
plt.show()
print("\n\n")
```

```python
# Dynamic Solutions to multiple equations
## What if we have multiple sets of two equations that intersect at one point, i.e. have one solution
## When we scale the original two equations using the each other, the new equations will have the same solution


eq1 = np.array([1, 2, 1])
eq2 = np.array([2, 1, 4])

for i in range(7):
    # new equations
    eq1_new = eq1 + np.random.randn(1) * eq2  # scaling using eq2
    eq2_new = eq2 + np.random.randn(1) * eq1  # scaling using eq1
    
    # plotting the solutions
    # using values of x from -5 to 5
    y = ([eq1_new[1] * -5, eq1_new[1] * 5] + eq1_new[2]) / eq1_new[0]    # solution for eq1
    plt.plot([-5, 5], y)

    # using values of x from -5 to 5
    y = ([eq2_new[1] * -5, eq2_new[1] * 5] + eq2_new[2]) / eq2_new[0]    # solution for eq2
    plt.plot([-5, 5], y)
    plt.axis([-5, 5, -5, 5])
    
    
    display.clear_output(wait = True)
    display.display(plt.gcf())
    time.sleep(0.3)

plt.show()
print("\n\n")
```

```python
## Equations with 3 variables
# Equation form: az = bx + cy + d
# a, b, c, d: coefficients
eq3 = np.array([1, 2, 3, -1])
eq4 = np.array([2, 1, 3, 3])


# plotting the solutions
figure = plt.figure(figsize=(18, 9))
ax = figure.gca(projection = '3d')


# using values of x from -3 to 3
y = ([eq3[1] * -3, eq3[1] * 3] + eq3[3]) / eq3[0]    # solution for eq1
# using values of y from -3 to 3
z = ([eq3[2] * -3, eq3[2] * 3] + eq3[3]) / eq3[0]    # solution for eq1
ax.plot([-3, 3], y, z)

# using values of x from -3 to 3
y = ([eq4[1] * -3, eq4[1] * 3] + eq4[3]) / eq4[0]    # solution for eq2
# using values of y from -3 to 3
z = ([eq4[2] * -3, eq4[2] * 3] + eq4[3]) / eq4[0]    # solution for eq2
ax.plot([-3, 3], y, z)

ax.set_xlim3d(-10, 10)
ax.set_ylim3d(-10, 10)
ax.set_zlim3d(-10, 10)

plt.show()
print("\n\n")
```

```python
# Dynamic Solutions to multiple equations
## What if we have multiple sets of three equations that intersect at one point, i.e. have one solution
## When we scale the original three equations using the each other, the new equations will have the same solution

eq3 = np.array([1, 2, 3, -1])
eq4 = np.array([2, 1, 3, 3])

# plotting the solutions
figure = plt.figure(figsize=(18, 9))
ax = figure.gca(projection = '3d')

for i in range(7):
    
    eq3_new = eq3 + (np.random.randn(1) * eq4)  # scaling using eq4
    eq4_new = eq4 + (np.random.randn(1) * eq3)  # scaling using eq3
    
    # using values of x from -3 to 3
    y = ([eq3_new[1] * -3, eq3_new[1] * 3] + eq3_new[3]) / eq3_new[0]    # solution for eq1
    # using values of y from -3 to 3
    z = ([eq3_new[2] * -3, eq3_new[2] * 3] + eq3_new[3]) / eq3_new[0]    # solution for eq1
    ax.plot([-3, 3], y, z)

    # using values of x from -3 to 3
    y = ([eq4_new[1] * -3, eq4_new[1] * 3] + eq4_new[3]) / eq4_new[0]    # solution for eq2
    # using values of y from -3 to 3
    z = ([eq4_new[2] * -3, eq4_new[2] * 3] + eq4_new[3]) / eq4_new[0]    # solution for eq2
    ax.plot([-3, 3], y, z)

    ax.set_xlim3d(-5, 7)
    ax.set_ylim3d(-5, 7)
    ax.set_zlim3d(-5, 7)
    
    display.clear_output(wait = True)
    display.display(plt.gcf())
    time.sleep(0.3)

plt.show()
print("\n\n")
```

## Solving the equations using matrices (reduced row Echelon form)

```python
# Sympy module has a Matrix class
# after converting the numpy array into Matrix, we can use the function rref (Gaussian elimination) defined in the Sympy module

# Previous equation form: ay = bx + c
# eq1 = np.array([1, 2, 1])
# eq2 = np.array([2, 1, 4])

# Equation: ay - bx = c  ==>  Martix form: [[a1, -b1], [[a2, -b2]] * [[y], [x]] = [[c1], [c2]]

mat1 = Matrix([[1, -2, 1], [2, -1, 4]])  # y and x coefficients concatenated with the constant vector


# We apply rref on the matrix [[a1, b1, -c1], [a2, b2, -c2]] and convert it into an echelon form [[1, 0, z1], [0, 1, z2]]

rref1 = mat1.rref()
print(np.array(rref1[0]), "\n\n")

print("y =\t", rref1[0][0, 2])
print("x =\t", rref1[0][1, 2])
print("\n\n")
```
