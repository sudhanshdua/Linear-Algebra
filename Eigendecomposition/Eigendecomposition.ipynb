## Finding Eigendecomposition

##### Eigendecompisition is only defined for square matrices, where as singular-value decomposition is defined for any n x m matrix

<br/>

##### When a matrix M (m x m) is multiplied by a vector v (m x 1), the resultant u is also a vector (m x 1),
##### the operation of multiplication with a matrix often rotates the vector (direction) and changes it's magnitude.
##### But,
##### there are some combinations of matrix M and vector v, for which the vector u produced by their multiplication has no rotation,
##### => vector v and the vector u have the same directions, maybe different magnitudes
##### => vector u is in the same 1-D sub-space as vector v
##### => vector u is just a scaled version of vector v
<br/>

####      $M . v = u$                   
#####          or 
####      $M . v = $$\ \alpha$$ . v$$\;\;\;\;\;\;\;(an\; entire\; matrix\; M\; is\; acting\; like\; a\; single\; number)$$ $
#####          or
####  $M . v $$\ - $$\ \alpha$$ . v = 0$
#####          or
####  $(M - $$\ \alpha$$ . I) . v = 0$$\;\;\;\;\;\;\;(for\; a\; non-zero\; vector\; v)$$ $
#####          or
####  $|M - $$\ \alpha$$ . I| = 0$$ $
<br/>

#####  If the above equation is true for a combination of a matrix and a vector, 
#####  then, 
#####       v --> Eigenvector, and 
#####  $\alpha$ --> Eignevalue
<br/>

##### Note: For any given matrix M, if there are two different eigenvalues, then their corresponding eigenvectors are also different
<br/>
<br/>

## Finding Eigenvectors

```python
# matrix
M = [[1, 5], [2, 4]]

# extracting the eigen values
eig_vals = np.linalg.eig(M)
print(eig_vals[0], "\n\n")


# vectors
vec1 = np.array([1, 1])   #   <--- is actually an eigen vector for matrix M
vec2 = np.random.randn(2, 1)  
vec2 = vec2 / np.linalg.norm(vec2) # making vector 2 a unit vector


# computing Mv
M_v_1 = M @ vec1
M_v_2 = M @ vec2

# plotting the vectors and Mv
plt.plot([0, vec1[0]], [0, vec1[1]], 'g', label='vector v1')
plt.plot([0, M_v_1[0]], [0, M_v_1[1]],'g--', label = 'vector u1')
plt.plot([0, vec2[0]], [0, vec2[1]], 'k', label= 'vector v2')
plt.plot([0, M_v_2[0]], [0, M_v_2[1]],'k--', label ='vector u2')
plt.legend()
plt.axis([-8, 8, -8, 8])
plt.show()
print("\n\n")
```

## Interesting property of a diagonal matrix

```python
# diagonal matrix
M = np.diag([1234, -5678])
print(M, "\n\n")
print(np.linalg.eig(M)[0], "\n\n")

## The diagonal elements are the eigen values for a diagonal matrix

M = np.diag(np.arange(3, 11))
print(M, "\n\n", np.linalg.eig(M)[0], "\n\n")
```

## Finding eigenvectors

```python
# matrix
M = [[1,2], [2,1]]

eig_vals, eig_vecs = np.linalg.eig(M)
print(eig_vals, "\n\n")
print(eig_vecs, "\n\n")

## python returns the normalized version of eigenvectors

# compute the norm of each eigenvector
mag_eig_vec1 = np.sqrt(np.sum(np.square(eig_vecs[:, 0])) )
mag_eig_vec2 = np.sqrt(np.sum(np.square(eig_vecs[:, 1])) )
print(mag_eig_vec1, mag_eig_vec2, "\n\n")

# plot
plt.plot([0, eig_vecs[0, 0]], [0, eig_vecs[1,0]], 'r', label='vec 1')
plt.plot([0, eig_vecs[0, 1]], [0, eig_vecs[1,1]], 'k', label='vec 2')

plt.axis([-1,1,-1,1])
plt.legend()
plt.show()
print("\n\n")
```

## Diagonalization

```python
# creating a symmetric matrix (symmetric matrices have real eigen values --> easy to visualise)
M = np.random.randn(5, 5)
M = M.T @ M

# eigendecomposition
eig_vals, eig_vecs= np.linalg.eig(M)

## reconstrucing the matrix M using the eigenvalues and eigenvectors
## keeping all the eigen values on the diagonals of a diagonal matrix (i.e all other values = 0) in the matrix Lambda
## M = V . Lambda . V^-1, where V is the matrix formed by all the eigenvectors
M2 = eig_vecs @ np.diag(eig_vals) @ np.linalg.inv(eig_vecs)

# plotting
plt.subplot(121)
plt.imshow(M)
plt.axis('off')
plt.title('M')
plt.subplot(122)
plt.imshow(M2)
plt.axis('off')
plt.title('$V\Lambda V^{-1}$')
plt.show()
print("\n\n")
```

## Matrix powers via Diagonalization

###### $M = V\Lambda V^{-1}$
###### $M^3 = M . M . M$
###### => $ M^3 = (V\Lambda V^{-1}) . (V\Lambda V^{-1}) . (V\Lambda V^{-1})$
###### => $ M^3 = V \; \Lambda \; (V^{-1} . V) \; \Lambda \; (V^{-1} . V) \; \Lambda \; V^{-1}$
###### => $ M^3 = V \; \Lambda \; \Lambda \; \Lambda \; V^{-1}$
###### => $ M^3 = V \; \Lambda^{3} \; V^{-1}$
<br/>


```python
M = np.random.randn(2, 2)

# compute matrix power directly
print(np.linalg.matrix_power(M, 3), "\n\n")     # M @ M @ M

# matrix power via eigendecomposition
D, V = np.linalg.eig(M)
D = np.diag(D)

# reproducing M^3 via diagonalization
print(V @ np.linalg.matrix_power(D, 3) @ np.linalg.inv(V), "\n\n")
```

```python
# Computing the eigenvalues of M and M^3

# create a symmetric matrix
M = np.random.rand(3, 3)
M = M @ M.T

D, V   = np.linalg.eig(M)
D3, V3 = np.linalg.eig(M @ M @ M)

print(V, "\n\n", V3, "\n\n")

# plotting
plt.figure(figsize= (12, 6))
plt.subplot(221)
plt.imshow(V)
plt.axis('off')
plt.title('eigenvectors of matrix M')

plt.subplot(223)
plt.imshow(V3)
plt.axis('off')
plt.title('eigenvectors of matrix M^3')

plt.show()
print("\n\n")
## the signs for the eigenvectors for both the matrices might be different


## If we cube the eigenvalues of matrix M, we will get the eigenvalues of matrix M^3
print("Eigenvalues of matrix M:\t", D, "\n\n")
print("Eigenvalues of matrix M^3:\t", D3, "\n\n")
print("(Eigenvalues of matrix M)^3:\t", D**3, "\n\n")


# plotting the eigenvectors and the eigenvalues of the matrices M and M^3

fig = plt.figure()
ax = fig.gca(projection = '3d')

# plot the eigenvectors of M
ax.plot([0, V[0, 0]], [0, V[1, 0]], [0, V[2, 0]], 'b')
ax.plot([0, V[0, 1]], [0, V[1, 1]], [0, V[2, 1]], 'g')
ax.plot([0, V[0, 2]], [0, V[1, 2]], [0, V[2, 2]], 'r')

# plot the eigenvectors of M^3
ax.plot([0, V3[0, 0]], [0, V3[1, 0]], [0, V3[2, 0]], 'b--')
ax.plot([0, V3[0, 1]], [0, V3[1, 1]], [0, V3[2, 1]], 'g--')
ax.plot([0, V3[0, 2]], [0, V3[1, 2]], [0, V3[2, 2]], 'r--')

plt.title("Eigenvectors")
plt.show()
print("\n\n")

# plot the eigenvalues
plt.plot([1, 2, 3], D, 'bs-', label= 'M')
plt.plot([1.1, 2.1,3.1], D3, 'ro-', label= 'M^3')
plt.title('Eigenvalues')
plt.legend()
plt.show()
print("\n\n")
```

## Eigendecomposition of Symmetric matrices

##### *Eigenvectors of a symmetric matrix are pair-wise orthogonal (if they have distict eigenvalues)*

###### Since, two different eigenvalues must have different eigenvectors, we assume that $\alpha_1\;!=\;\alpha_2$

###### $\alpha_1\;v_1^{T}\;v_2$
###### $ = (M.\;v_1)^{T}\;v_2$ 
###### $ = v_1^{T}\;.M^{T}\;v_2$ 
###### $ = v_1^{T}\;.M\;v_2\;\;\;\;\;\;(M\;=\;M^{T}$, since M is symmetric)
###### $ = v_1^{T}\;.\alpha_2\;v_2$
###### $ = \alpha_2\;v_1^{T}\;v_2$
<br/>

###### => $\;\;\;\alpha_1\;v_1^{T}\;v_2 = \alpha_2\;v_1^{T}\;v_2$
###### => $\;\;\;(\alpha_1\;v_1^{T}\;v_2) - (\alpha_2\;v_1^{T}\;v_2) = 0$
###### => $\;\;\;(\alpha_1 - \alpha_2).\;v_1^{T}\;v_2 = 0$
###### => $\;\;\;v_1^{T}\;v_2 = 0\;\;\;\;\;\;\;\;\;(Since,\; \alpha_1 != \alpha_2)$

##### Note: If the symmetric matrix has repeated eigenvalues, then $\alpha_1 = \alpha_2$, and the eigenvectors doesn't necessarily have to be orthogonal
<br/>
<br/>

---
### For a symmetric matrix, the result of the eigenvector matrix times its transpose is an Identity matrix

```python
# create a random matrix
M = np.random.randn(20, 20)

# making it symmetric (additive method)
M = M + M.T

# diagonalize it (eigendecomposition)
eig_vals, eig_vecs = np.linalg.eig(M)
print("\nEigenvectors:\n", np.round(eig_vecs, 3), "\n\n")

# magnitudes of each vector
print("Magnitude of the eigenvectors:\n", np.sqrt(sum(eig_vecs**2)), "\n\n")

# plotting
plt.imshow(M)
plt.axis('off')
plt.title('M')
plt.show()

plt.imshow(eig_vecs)
plt.axis('off')
plt.title('Eigenvectors')
plt.show()

plt.imshow(eig_vecs @ eig_vecs.T)
plt.axis('off')
plt.title('$V\;V^{T}$')
plt.show()
print("\n\n")
```

##### Does the trace of a matrix = sum of its eigenvalues?
##### Does the determinant of a matrix = product of its eigenvalues?

```python
## Case 1: Full rank square matrix
M1 = np.random.randn(7, 7)
trace1 = np.matrix.trace(M1)
determinant1 = np.linalg.det(M1)
eig_vals1 = np.linalg.eig(M1)[0]
print("Full rank matrix:")
print(trace1, "\t", sum(eig_vals1))
print(np.round(determinant1, 6), "\t\t", np.round(np.prod(eig_vals1), 6), "\n\n")

## Case 2: Reduced rank matrix (Rank = 5)
M2 = np.random.randn(7, 5) @ np.random.randn(5, 7) 
trace2 = np.matrix.trace(M2)
determinant2 = np.linalg.det(M2)
eig_vals2 = np.linalg.eig(M2)[0]
print("Reduced rank matrix:")
print(trace2, "\t", sum(eig_vals2))
print(np.round(determinant2, 6), "\t\t\t", np.round(np.prod(eig_vals2), 6), "\n\n")
```
