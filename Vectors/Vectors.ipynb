# Vectors

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as axes3D
```

## Plotting

```python
# 2-D vector
v1 = [2, 4]

# 3-D vector
v2 = [1, 2, 3]

# Vector transpose (i.e. from Row vector to a Column vector and vice versa)
v2_t = np.transpose(v2)

# v2 is a column vector and v2_t is the row vector
print(v2, v2_t)
```

## Numerical Operations on Vectors

```python
# Vector addition

# two column vectors
v3 = np.array([1, 3])
v4 = np.array([4, -2])

v5 = v3 + v4

# Plotting all of them together (showing how the gemoetric vector addition)
#
# line staring at (0,0) and ending at (v3[0], v3[1])
plt.plot([0, v3[0]],[0, v3[1]], 'b', label ='v3')
# Line starting at (v3[0], v3[1]) and ending at ( (v3[0] + v4[0]), (v3[1] + v4[1]) )
plt.plot([0, v4[0]] + v3[0], [0, v4[1]] + v3[1], 'r', label ='v4')
# line staring at (0,0) and ending at (v5[0], v5[1])
plt.plot([0, v5[0]], [0, v5[1]], 'y', label ='v3 + v4')

plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6))
plt.grid()
plt.show()
```

```python
# Vector subtraction
v6 = v3 - v4

# Plotting all of them together (showing how the gemoetric vector addition)
#
# line staring at (0,0) and ending at (v3[0], v3[1])
plt.plot([0, v3[0]],[0, v3[1]], 'b', label ='v3')
# Line starting at (v3[0], v3[1]) and ending at ( (v3[0] - v4[0]), (v3[1] - v4[1]) )
plt.plot([0, -v4[0]] + v3[0], [0, -v4[1]] + v3[1], 'r', label ='v4')
# line staring at (0,0) and ending at (v5[0], v5[1])
plt.plot([0, v6[0]], [0, v6[1]], 'y', label = 'v3 - v4')

plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6))
plt.grid()
plt.show()
```

```python
# Vector mutilpication with a scalar

# column vector 
v7 = np.array([-1, 1])

# scalars
l1 = 7
l2 = -2
# new vector (scaled v7)
v8 = v7 * l1
v9 = v7 * l2

# Plotting them
plt.plot([0, v7[0]], [0, v7[1]], 'b', label='$v_7$')
plt.plot([0, v8[0]],[0, v8[1]],'r:',label='$\lambda_1 * v_7$')
plt.plot([0, v9[0]],[0, v9[1]],'g:',label='$\lambda_2 * v_7$')

plt.legend()
plt.axis('square')

# dynamic axis limits (uses the maximum x- and y- values from the vectors)
axlim = max([max(abs(v7)), max(abs(v8)), max(abs(v9))]) * 1.5
plt.axis((-axlim, axlim, -axlim, axlim))

plt.grid()
plt.show()
```

## Dot Product (Algebraic)

```python
# Vector mutilpication with another vector a.k.a the Dot Product

v10 = np.array([ 1, 2, 3, 4, 5, 6 ])
v11 = np.array([ 0, 2, 4, 6, 8, 10])

# many ways to compute the dot product
# method 1
# the multiply method gives a row vector (1 x 6)
dp_1 = sum(np.multiply(v10, v11))

# method 2
dp_2 = np.dot(v10, v11)

# method 3
dp_3 = np.matmul(v10, v11)

print(dp_1,dp_2,dp_3)
```

## Dot product properties:

```python
## 1) Distributive property:    Is   A.(B + C) = A.B + A.C    ??
## 2) Associative property:     Is   A.(B.C) = (A.B).C        ??


# Creating random vectors
n = 10
A = np.random.randn(n)
B = np.random.randn(n)
C = np.random.randn(n)

# A.(B + C)
dp_5 = np.dot(A, B + C)
# A.B + A.C
dp_6 = np.dot(A, B) + np.dot(A, C)
print(round(dp_5 - dp_6, 5), "\n\n")        # (dp_5 - dp_6) = 0    => Distributive property always hold


# A.(B.C)
dp_7 = np.dot(A, np.dot(B, C))
# (A.B).C
dp_8 = np.dot(np.dot(A, B), C)
print(dp_7 - dp_8, "\n\n")                  # (dp_7 - dp_8) != 0   => Associative property doesn't hold



# There are two cases when the Associateive property holds 
## case 1) if one of the vectors is a zero vector
C = np.zeros(n)
# A.(B.C)
dp_9 = np.dot(A, np.dot(B, C))
# (A.B).C
dp_10 = np.dot(np.dot(A, B), C)
print(dp_9 - dp_10,"\n\n")                  # (dp_9 - dp_10) = 0   => Associative property holds

## case 2) if A == B == C
A = np.array([1, 2, 3])
B = np.array([1, 2, 3])
C = np.array([1, 2, 3])
# A.(B.C)
dp_11 = np.dot(A, np.dot(B, C))
# (A.B).C
dp_12 = np.dot(np.dot(A, B), C)
print(dp_11 - dp_12)                        # (dp_11 - dp_12) = 0   => Associative property holds
```

## Length of a Vector

```python
# Geometrically, a vector A[3, 4] can be seen as a line staring at (0,0) and ending at (3, 4)
# The length of such a line = ((x-0)^2 + (y-0)^2)^0.5

# Mathematically, length = (A'A)^0.5


v12 = np.array([1, 2, 3, 4, 5])

length1 = np.sqrt(np.dot(v12, v12))

# alternative method: take the norm
length2 = np.linalg.norm(v12)

print(length1, length2)
```

## Dot Product (Geometric)

```python
# two vectors
v13 = np.array([ 0,  2, -4 ])
v14 = np.array([ 2, -4, -1 ])

# computing the angle (in radians) between two vectors
## dot product of two vectors a, b = |a|.|b|.cos(Theta) => Theta = cos_inverse(a.b / (|a|.|b|)), 
## where |v| is the length of the vector v
angle = np.arccos(np.dot(v13, v14) / (np.linalg.norm(v13) * np.linalg.norm(v14)))


# plotting the vectors
fig = plt.figure(figsize = (14, 7))
ax = fig.gca(projection='3d')
ax.plot([0, v13[0]], [0, v13[1]], [0, v13[2]], 'b')
ax.plot([0, v14[0]], [0, v14[1]], [0, v14[2]], 'r')

plt.axis((-5, 5, -5, 5))
plt.title("Angle between the vectors: %f radians" %angle)
plt.show()
```

## Vector Outer product

```python
v15 = np.array([  1, 2, 3 ]) # 1x3 vector
v16 = np.array([ -1, 0, 1 ]) # 1X3 vector

dp_13 = np.dot(v15, v16) # scalar

# outer product
op_1 = np.outer(v15, v16) # 3x3 matrix

print("Dot Product:\t", dp_13, "\n\n", "Outer Product:\n", op_1)
```

## Vector Cross Product

```python
v17  = [1, 2, 0] # in the x-y plane 
v18  = [4, 3, 0] # in the x-y plane

# cross product will be in the z-axis
cp_1 = np.cross(v17, v18)

print(cp_1, "\n\n")


fig = plt.figure(figsize = (14, 7))
ax = fig.gca(projection='3d')

# drawing the plane defined by the span of v17 and v18
xx, yy = np.meshgrid(np.linspace(-10, 10, 11), np.linspace(-10, 10, 11))
z1 = (-cp_1[0] * xx - cp_1[1] * yy) / cp_1[2]
ax.plot_surface(xx, yy, z1, alpha=.5)

## plot the two vectors
ax.plot([0, v17[0]], [0, v17[1]], [0, v17[2]], 'k')
ax.plot([0, v18[0]], [0, v18[1]], [0, v18[2]], 'k')
ax.plot([0, cp_1[0]], [0, cp_1[1]], [0, cp_1[2]],'r')


ax.view_init(azim = 150, elev = 45)
plt.show()
```

### Conjugate Transpose


```python
# complex vector
v19 = complex(3, 4)    # 3 + 4j, where j = (-1) ^ 0.5
print("v19 =", v19)

# length of this vector in a complex plane = ( x^2 + y^2) ^ 0.5
print("v19 length =", np.linalg.norm(v19))   # 5.0


# For a complex number a + bj, it's conjugate is a - bj
print("v19 conjugate =", np.conjugate(v19), "\n")


# complex vector * (Complex vector).T != 25
print(np.transpose(v19) * v19)


# complex vector * ( conjugate of the complex vector ) =>  real number = (length of the complex number) ^ 2
print(np.conjugate(v19) * v19, "\n")


# vector of complex vectors
v20 = np.array([3, 4j, 5 + 12j, complex(-3, 4), complex(5, -12)])
print(v20)
print(v20.T, "\n") # transpose

print(np.conjugate(v20)) # conjugate
print(np.conjugate(v20).T) # transpose of a complex conjugate
```

## Unit vector

```python
v21 = np.array([3, 4])
unit_v1 = v21 / (np.linalg.norm(v21))  # unit vector 1 = v21 / length of v21, we get a vector in the direction of v21, with a magnitude of 1
print(unit_v1)

# plotting them
plt.plot([0, v21[0]], [0, v21[1]], 'b', label = "vector")
plt.plot([0, unit_v1[0]], [0, unit_v1[1]], 'gold', label = 'unit vector', linewidth = 3)
plt.axis('square')
plt.axis((-5, 5, -5, 5))
plt.grid()
plt.legend()
plt.show()
```
